# **Calculate accidents in day time and night time in 2014.**

```python
dayTime2014 = df['03-600hrs - Night - 2014']+ df['06-900hrs - Day - 2014'] + \
              df['09-1200hrs - Day - 2014']+df['12-1500hrs - Day - 2014']+df['15-1800hrs - Day - 2014']
nightTime2014 = df['00-300hrs - Night - 2014'] + df['18-2100hrs - Night - 2014']+df['21-2400hrs - Night - 2014']
```

```python
dayTime2014
```

```python
nightTime2014
```

```python
plt.rcParams.update({'font.size'  : 18})
fig = plt.figure(figsize = (20,10))

ax = fig.add_axes([0,0,1,1])

ax.bar(df.index, dayTime2014, label = 'Day Time')
ax.bar(df.index,nightTime2014, label = 'Night Time')

plt.title("Line chart showing flow of number of accidents happening at DAY TIME and NIGHT TIME for 2014.")
plt.xlabel("States/UTs")
plt.ylabel("Count of Accidents")
plt.xticks(rotation = 90)
plt.legend( loc = 'best')
plt.show()
```

# **Total day and night time accidents in 2014 and 2016**

```python
dayTime2016 = df['03-600hrs - (Night) - 2016']+ df['06-900hrs - (Day) - 2016'] + \
              df['09-1200hrs - (Day) - 2016']+df['12-1500hrs - (Day) - 2016']+df['15-1800hrs - (Day) - 2016']
nightTime2016 = df['00-300hrs - (Night) - 2016'] + df['18-2100hrs - (Night) - 2016']+df['21-2400hrs - (Night) - 2016']
```

```python
dayTime2014
```

```python
dayTime2016
```

```python
dayTimeTotal =  dayTime2014 + dayTime2016
```

```python
fig = plt.figure(figsize = (20,10))

ax = fig.add_axes([0,0,1,1])
UT = np.arange(36)

ax.bar(UT,dayTime2014, width = 0.2, label = 'Day Time 2014')
ax.bar(UT+0.2,dayTime2016, width = 0.2,label = 'Day Time 2016')

plt.title("Line chart showing flow of number of accidents happening at DAY TIME and NIGHT TIME for 2014.")
plt.xlabel("States/UTs")
plt.ylabel("Count of Accidents")
plt.xticks(UT, df.index,rotation = 90)
plt.legend( loc = 'best')
plt.show()
```
